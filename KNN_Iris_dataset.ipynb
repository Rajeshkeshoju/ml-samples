```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

import pandas as pd
```

```python
irisDataCols = load_iris()
```

```python
irisData = pd.DataFrame(irisDataCols.data)
irisDataClassLabels = pd.DataFrame(irisDataCols.target)
```

```python
columns = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]
classLabels = ["Setosa", "Versicolour", "Virginica"]
```

```python
irisData.columns = columns
irisData
```

```python
irisDataClassLabels.columns = ["Species"]
irisDataClassLabels["Species"] = irisDataClassLabels["Species"].replace(classLabels, [0,1,2])
irisDataClassLabels
```

```python
x = irisDataCols.data
y = irisDataCols.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

knn = KNeighborsClassifier() # default is KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

# Printing Training data with class label from dataset
tddf=pd.DataFrame(x_train)
tddf.columns = columns
tddf["Species"] = y_train
tddf
```

```python
# Test data from dataset
# dataset test data classification

test = pd.DataFrame(x_test)
test.columns = columns

# Classifying test data
classLabel = knn.predict(test)
# Adding Species column as resulting class labels from KNN model
test['Species'] = classLabel

# Representing test data and appending result classified labels from KNN model
test['Species'] = test['Species'].replace([0,1,2], classLabels)
test
```

```python
# Custom test data classification

newTestData = [[1.9, 9.4, 1.6, 7.8], [7.5, 2.3, 5.1, 8.8], [1.2, 3.4, 5.5, 6.6], [9.9, 3.2, 2.3, 4.3], [1.9, 5.6, 3.3, 2.0]]
test = pd.DataFrame(newTestData)
test.columns = columns

# Classifying test data
classLabel = knn.predict(test)
# Adding Species column as resulting class labels from KNN model
test['Species'] = classLabel

# Representing test data and appending result classified labels from KNN model
test['Species'] = test['Species'].replace([0,1,2], classLabels)
test
```
